## #save DATA to database

```python
# define connection and cursor
connexion = sqlite3.connect("database.db")
cursor = connexion.cursor()
```

```python

cursor.execute('''
        CREATE TABLE IF NOT EXISTS Table_B3
        (coilId INTEGER,
        lengthCoordinate REAL,
        B3 REAL,
        CONSTRAINT field_pk PRIMARY KEY (field_B3)
        )
        ''')
print("table1 created")
connexion.commit()

cursor.execute('''
        CREATE TABLE IF NOT EXISTS Table_B4
        (coilId INTEGER,
        lengthCoordinate REAL,
        B4 REAL,
        CONSTRAINT field_pk PRIMARY KEY (field_B4)
        )
        ''')
print("table2 created")
connexion.commit()

cursor.execute('''
        CREATE TABLE IF NOT EXISTS Table_B5
        (coilId INTEGER,
        lengthCoordinate REAL,
        B5 REAL,
        CONSTRAINT field_pk PRIMARY KEY (field_B5)
        )
        ''')
print("table3 created")
connexion.commit()

cursor.execute('''
        CREATE TABLE IF NOT EXISTS Coil_Information
        (coil INTEGER,
        furnace_Number INTEGER,
        analyse TEXT,
        Hardness_1 INTEGER,
        Hardness_2 INTEGER,
        Width REAL,
        Temp_before_finishing_mill INTEGER,
        Temp_after_finishing_mill REAL,
        Thickness REAL,
        Thickness_profile REAL,
        C INTEGER,
        Si INTEGER,
        Mn INTEGER,
        Nb INTEGER,
        P INTEGER,
        S INTEGER,
        Al INTEGER,
        M_Al INTEGER,
        B INTEGER,
        N INTEGER,
        Ti INTEGER,
        Cr INTEGER,
        V INTEGER,
        Mo INTEGER,
        CONSTRAINT coil_pk PRIMARY KEY (coil)
        )
        ''')
print("Coil_Information created")
connexion.commit()


```

```python
file = pd.read_csv('B3.csv')

for row in file:
    cursor.execute('''
        INSERT INTO Table_B3(coilId, lengthCoordinate, B3)
        VALUES(?,?,?)
        ''', row)
connexion.commit()
```

```python
file2 = pd.read_csv('B4.csv', header=0)
for row in file2:
    cursor.execute('''
        INSERT INTO Table_B5(coilId, lengthCoordinate, B5)
        VALUES(?,?,?)
        ''', row)
connexion.commit()
```

```python
file3 = open('B5.csv', 'r')
reader_B5 = csv.reader(file3)
for row in reader_B5:
    cursor.execute('''
        INSERT INTO Table_B5(coilId, lengthCoordinate, B5)
        VALUES(?,?,?)
        ''', row)
connexion.commit()
```
