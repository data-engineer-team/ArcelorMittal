## clean up B3_B4

```python
df_b3_b4 = pd.read_csv('dumpRawDataB3_B4.csv')
df_b3_b4.shape
```

```python
df_b3_b4.columns
```

```python
df_b3_b4['lengthCoordinate'] = pd.Series(dtype='float64')
df_b3_b4['B3'] = pd.Series(dtype='float64')
df_b3_b4['B4'] = pd.Series(dtype='float64')

df_b3_b4
```

```python
df_b3_b4.info()
```

```python
df_b3_b4[['B3_Time_Based', 'B4_Time_Based']] = df_b3_b4[['B3_Time_Based', 'B4_Time_Based']].fillna(0)
df_b3_b4
```

```python
for i in range(len(df_b3_b4)):
    df_b3_b4['B3'][i] = df_b3_b4['B3_Time_Based'][i] * 0.1
    df_b3_b4['B4'][i] = df_b3_b4['B4_Time_Based'][i] * 0.1
```

```python
df_b3_b4
```

```python
# get the lengthCoordinate
for i in range (len(df_b3_b4)):
    if i == 1:
        df_b3_b4['lengthCoordinate'] =  df_b3_b4['Snelheid_EW'] * 0.001 * 0.04
    elif i == 0:
        df_b3_b4['lengthCoordinate'] = 0
    else:
        df_b3_b4['lengthCoordinate'][i] = df_b3_b4['lengthCoordinate'][i-1] + df_b3_b4['Snelheid_EW'][i-1] * 0.001 * 0.04
    
```

```python
df_b3_b4.fillna({'B3_Abs_Plaus_Time': 0, 'B4_Abs_Plaus_Time': 0}, inplace=True)
```

```python
df_b3_b4
```

```python
df_b3_b4.dtypes
```

## cleaning up B5

```python
df_b5 = pd.read_csv('dumpRawDataB5.csv')
df_b5.shape
```

```python
df_b5['lengthCoordinate'] = pd.Series(dtype='float64')
df_b5['B5'] = pd.Series(dtype='float64')
df_b5
```

```python
df_b5 = df_b5.fillna(0)

df_b5.info()
```

```python
df_b5
```

```python
for i in range(len(df_b5)):
    if i == 1:
        df_b5['lengthCoordinate'] = df_b5['Snelheid_HA'] * 0.001 * 0.04
    elif i == 0:
        df_b5['lengthCoordinate'] = 0
    else:
        df_b5['lengthCoordinate'][i] = df_b5['lengthCoordinate'][i-1] + df_b5['Snelheid_HA'][i-1] * 0.001 * 0.04


    
    
    
```

```python
df_b5
```

```python
df_b5.at[0, 'lengthCoordinate'] = 0
```

```python
df_b5_new = df_b5[['_FileId', '_SegmentNr', 'Snelheid_HA', 'lengthCoordinate','B5', 'B5_Abs_Plaus_Time', 'B5_Time_Based']]
```

```python
for i in range(len(df_b5_new)):
    df_b5_new['B5'][i] = df_b5_new['B5_Time_Based'][i] * 0.1
```

```python
df_b5_new = df_b5_new.rename(columns={'lengthCoordinate': 'lengthCoordinateB5'}, inplace=False)
```

```python
df_b5_new.to_csv('dataB5.csv', index=False)
```
